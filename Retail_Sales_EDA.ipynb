```python
#required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import json
```

```python
#reading the dataset
#For Behaviour data
path_behv = '/content/drive/MyDrive/Retail_Sales_ABADS/behaviour.json'

with open(path_behv, 'r') as file:
    data_behv = json.load(file)

df_behv = pd.DataFrame()

for entry in data_behv:
    for key, value in entry.items():
        row_data = {"ID": key, **value}
        df_behv = df_behv.append(row_data, ignore_index=True)

#For Campaign data
path_camp = '/content/drive/MyDrive/Retail_Sales_ABADS/campaign.json'
with open(path_camp, 'r') as file:
    data_camp = json.load(file)

df_camp = pd.DataFrame()

for entry in data_camp:
    for key, value in entry.items():
        row_data = {"ID": key, **value}
        df_camp = df_camp.append(row_data, ignore_index=True)

#For Demographics data
path_demo = "/content/drive/MyDrive/Retail_Sales_ABADS/demographics.txt"
df_demo = pd.read_csv(path_demo, sep='\t')

```
